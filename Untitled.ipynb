```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dirección de las imagenes
img_dir = "/home/alejoxbg/parcial/PyTorch-YOLOv3/data/custom/images/"
list_img=sorted(os.listdir(img_dir))[0:]

#dirección de las etiquetas
ann_dir = "data_object_label_2/training/label_2/"
list_ann=sorted(os.listdir(ann_dir))[1:]


##prueba para verificar la lectura
x=0
img_path =img_dir+list_img[x]
img=cv2.imread(img_path)
ann_path=ann_dir+list_ann[x]
gt=np.loadtxt(ann_path, delimiter =" ",dtype=str)
```

```python
#En esta celda se va a encontrar todas las 

#se crean vectores vacios
clases=[]
gt=[]
n=0

#se recorren todas las anotaciones de la carpeta
for i in range(len(list_ann)):
    
    #se crea el nombre de la anotación correspondiente por cada iteración
    ann_path=ann_dir+list_ann[i]
   
    #se carga y se asigna la lectura del archivo txt a una variable llamada gt
    gt=np.loadtxt(ann_path, delimiter =" ",dtype=str)
    
    #se detecta si la anotación solo tiene 1 objeto detectado o más
    if len(np.loadtxt(ann_path, delimiter =" ",dtype=str).shape) > 1:
        n=0
        #si tiene mas de 1 detección se recorren todas las detecciones y se extrae la clase de cada dettección
        for l in range(np.loadtxt(ann_path, delimiter =" ",dtype=str).shape[0]):
            clases.append(np.loadtxt(ann_path, delimiter =" ",dtype=str)[0:][n][0])
            n=n+1
    else:
        #si tiene solo 1 detección se recorren todas las detecciones y se extrae la clase de cada dettección
            clases.append(np.loadtxt(ann_path, delimiter =" ",dtype=str)[0:][0])
                
    n=n+1
```

```python
#se crea un vector con las clases unicas
classes_name=np.unique(clases)

#se imprime el vector para poder colocarlo en el archivo classes.name en pytorchyolov2
for i in np.unique(clases):
    print(i)
```

```python
#se crea u diccionario donde se le asigna un valor a cada clase
classes_ids={cat:i for i, cat in enumerate(classes_name)}
```

```python
#se define la carpeta donde se guardará el archivo
save_in="/home/alejoxbg/parcial/PyTorch-YOLOv3/data/custom/labels/"

#se recorre todas las anotaciones
for annx in range(len(list_ann)):
    
    #se obtiene la dirección de la imagen
    img_path=img_dir+list_img[annx].replace('txt','jpg')
    
    #se carga la imagen con opencv
    img=cv2.imread(img_path)
    
    #se obtiene la dirección de la anotación correspondente
    ann_path=ann_dir+list_ann[annx]
    
    #se obtiene las dimenciones de la imagen
    h,w,c=img.shape
    
    #se carga y se asigna la lectura del archivo txt a una variable llamada gt
    gt=np.loadtxt(ann_path,delimiter=" ",dtype=str)
    
    
    gt_converted=[]

    #se detecta si la anotación solo tiene 1 objeto detectado o más
    if len(np.loadtxt(ann_path, delimiter =" ",dtype=str).shape) > 1:
        for i in gt:
            #si tiene mas de 1 detección se recorren todas las detecciones y se extrae la clase de cada dettección
            
            #en las siguientes lineas se obtienen las dimensiones de la bounding box y se pasa al formato de yolo
                width=i[6].astype(float)-i[4].astype(float)
                height=i[7].astype(float)-i[5].astype(float)
                center_x=width/2+i[4].astype(float)
                center_y=height/2+i[5].astype(float)
                gti=[classes_ids[i[0]]]+[center_x/w]+[center_y/h]+[width/w]+[height/h]
                gt_converted.append(gti)
    else:
        #se realiza lo mismo pero para el caso donde solo existe un detección
        width=gt[6].astype(float)-gt[4].astype(float)
        height=gt[7].astype(float)-gt[5].astype(float)
        center_x=width/2+gt[4].astype(float)
        center_y=height/2+gt[5].astype(float)
        gti=[classes_ids[gt[0]]]+[center_x/w]+[center_y/h]+[width/w]+[height/h]
        gt_converted.append(gti)

        
    #finalmente se convierte la lista a un tipo array para poder guardarlo
    final_get=np.array(gt_converted)
    
    #se guarda en el directorio que se seleccionó
    np.savetxt(save_in+list_ann[annx],final_get, delimiter = ' ',fmt='%s')
```
