```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_dir = "/home/alejoxbg/parcial/PyTorch-YOLOv3/data/custom/images/"
list_img=sorted(os.listdir(img_dir))[:]

ann_dir = "/home/alejoxbg/parcial/data_object_image_2/data_object_label_2/training/label_2/"
list_ann=sorted(os.listdir(ann_dir))[1:]

```

```python
ann_dir
```

```python
classes_name = ['Car', 'Van', 'Truck', 'Pedestrian', 'Person_sitting', 'Cyclist', 'Tram', 'Misc', 'DontCare']
for i in classes_name:
    print(i)
```

```python
classes_ids=classes_ids={cat:i for i, cat in enumerate(classes_name)}
classes_ids
```

```python
save_in="data_object_image_2/ground_truth/"
for annx in range(len(list_ann)):
    img_path=img_dir+list_img[annx].replace('txt','jpg')
    img=cv2.imread(img_path)    
    ann_path=ann_dir+list_ann[annx]
    h,w,c=img.shape
    gt=np.loadtxt(ann_path,delimiter=" ",dtype=str)
    gt_converted=[]
    if len(np.loadtxt(ann_path, delimiter =" ",dtype=str).shape) > 1:
        for i in gt:
            x1=i[4].astype(float)
            y1=i[5].astype(float)
            x2=i[6].astype(float)
            y2=i[7].astype(float)
            gti=[classes_name[int(classes_ids[i[0]])]]+[int(x1)]+[int(y1)]+[int(x2)]+[int(y2)]
            gt_converted.append(gti)
    else:
        x1=gt[4].astype(float)
        y1=gt[5].astype(float)
        x2=gt[6].astype(float)
        y2=gt[7].astype(float)
        gti=[classes_name[int(classes_ids[gt[0]])]]+[int(x1)]+[int(y1)]+[int(x2)]+[int(y2)]
        gt_converted.append(gti)
        #print(gt_converted)
    final_get=np.array(gt_converted)
    np.savetxt(save_in+list_ann[annx],final_get, delimiter = ' ',fmt='%s')
    
```

```python

x=0
img_path =img_dir+list_img[x]
img=cv2.imread(img_path)
ann_path='labels/'+list_ann[x]
h,w,c=img.shape
gt=np.loadtxt(ann_path,delimiter=" ",dtype=float)
#print(gt_converted)
if len(np.loadtxt(ann_path, delimiter =" ",dtype=str).shape) > 1:
    for i in gt:
        #print(i)
        cv2.circle(img, (int(i[1]*w) , int(i[2]*h)) ,int(i[3]*w/2),(0,255,0),2)
else:
    cv2.circle(img, (int(gt[1]*w) , int(gt[2]*h)) ,int(gt[3]*w/2),(0,255,0),2)
               
fig , ax=plt.subplots(figsize=(15,15))
plt.imshow(img)
```
